```python
import pandas
```

```python
pkidb = pandas.read_csv("PKIDB_CSV.csv")
chembl_id = pandas.read_csv("ChEMBL_ID.csv")
refined=pandas.read_csv("Refined_PKIDB.csv")
smiles = pandas.read_csv("SMiles.csv")
```

```python
#ADD SMILES 
pkidb_inn = pkidb.INN_Name
smiles_inn = smiles.INN_Name
smiles = smiles.join(pkidb.set_index('INN_Name'), on='INN_Name')
smiles = smiles.drop(columns = ['Canonical_Smiles_InChiKey'], axis=1)
```

```python
#ADD CHEMBL ID
chembl_inn = chembl_id.INN_Name
chembl_id = chembl_id.join(smiles.set_index('INN_Name'), on='INN_Name')
chembl_id = chembl_id.drop(columns = ['BrandName','Phase','Applicants','Links','LigID','pdbID','Type',
                                      'Indications','Targets','First_Approval','SC_Patent','Chirality',
                                      'FDA approved','Melting points (Â°C)','Synonyms',], axis=1)
image_list = []
for ID in chembl_id['INN_Name']:
    image_link = "http://www.icoa.fr/pkidb/static/img/mol/"+ID+".svg"
    image_list.append(image_link)
chembl_id["image_link"] = image_list
chembl_id = chembl_id.set_index("INN_Name")
#chembl_id.to_csv("Inhibitor_Data_Final.csv")
```

```python
#ADD TARGETS
refined_inn = refined.INN_Name
refined = refined.join(chembl_id.set_index('INN_Name'), on='INN_Name')
refined = refined.set_index("INN_Name")
#refined.to_csv("Target_Data_Final.csv")
```
