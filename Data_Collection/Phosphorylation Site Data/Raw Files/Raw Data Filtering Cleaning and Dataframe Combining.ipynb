```python
#Open both of our raw data files in pandas and converting them into panda dataframes
KinaseSubstrateDF = pd.read_excel('Kinase_Substrate_Dataset_UNEDITED.xlsm')
PhosphorylationSiteDF = pd.read_excel('Phosphorylation_site_dataset_UNEDITED.xlsm')
```

```python
###Filtering and Cleaning for Kinase_Substrate dataset###

#Filter data in the data frame to only include data where the Kinase Organism and Substrate Organism are both Human
HUMANfilteredKSDF = KinaseSubstrateDF[(KinaseSubstrateDF.KIN_ORGANISM=='human') & (KinaseSubstrateDF.SUB_ORGANISM=='human')]

#Remove the columns of data in the data frame that we do not need for our database
KSDFColumnsEdit = HUMANfilteredKSDF.drop(['KIN_ORGANISM', 'SUB_ORGANISM', 'IN_VIVO_RXN', 'IN_VITRO_RXN', 'CST_CAT#'],axis=1)

#Changing the column title for Neighbouring AAs so that if we want to manipulate it further in Python we can do without syntax errors
KSDFColumnCorrected = KSDFColumnsEdit.rename(columns = {'SITE_+/-7_AA': 'SITE_7_AA'}) 
```

```python
###Filtering and Cleaning for Phosphorylation_Dataset dataset###

#Filter data in the data frame to only include data where the Kinase Organism and Substrate Organism are both Human
HUMANfilteredPSDF = PhosphorylationSiteDF[PhosphorylationSiteDF.ORGANISM == 'human']

#As we only need this file for the Genomic Locations no further filtering and cleaning is needed.
```
